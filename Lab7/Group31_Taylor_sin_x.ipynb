```python
# Returns Min n, |x^{n}/n!| < epsilon and _{i=0}^{n} x^i/i!

import time

start = time.time()

x = 0.5
epsilon = 10**(-20)

xo = x

if x < 0 :
   x = - x
#int (delta)

delta = 1
sin_x = x
i = 1

while  epsilon <= abs(delta):
  #print(delta, sin_x)
  delta *= ((x**(2*i+1))/((2*i+1)*(2*i))) 
  sin_x += ((-1)**(i))*delta  
  i = i+1

if x != xo: 
  sin_x = 1 / sin_x
  x = xo

end = time.time()

print("n = ",i-1) 
print('x = ',x)
print('e_to_x  _{i=0}^'+str(i-1),'_({i=0}^(i)*((x**(2*i+1))/(2*i+1))! =',sin_x)
print("epsilon =","{:.0e}".format(epsilon))     
print('math.sin(x) = ',math.sin(x))
print('time = ',end - start)
```

```python
# Returns Min n, |Lagrange_Residual_n| < epsilon and _{i=0}^{n} x^i/i!

import math
import time

start = time.time()
x = 0.5
epsilon = 10**(-20)

sin_x = 0

print("n".rjust(10)," ","".center(21)," ","|Lagrange_Residual_n|".center(20)," ","epsilon".center(20))
error = 1
i = 0
while  epsilon <= abs(error):
  sin_x += ((-1)**(i))*((x**(2*i+1))/math.factorial(2*i+1))  
  if ( x >= 0 ):
     error = math.sin(x) * ((-1)**(i+1))*((x**(2*(i+1)+1))/math.factorial(2*(i+1)+1)) 
  else:
     error = ((-1)**(i+1))*((x**(2*(i+1)+1))/math.factorial(2*(i+1)+1))
  print(format(i, '10'),"  ", format(sin_x, '.17f'),"  ",format(error, '.17f'),"  ","{:.0e}".format(epsilon)) 
  i = i+1
end = time.time()  
print("n = ",i-1) 
print('x = ',x)
print('sin_of_x  _{i=1}^'+str(i),'[(-1)^('+str(i),')][(x^2*'+str(i),'+1/2*'+str(i),'+1)] =',sin_x)
print("epsilon =","{:.0e}".format(epsilon))   
print('math.sin(x) = ',math.sin(x))
print('time = ',end - start)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

values = np.arange(-5,5,0.1)
p_sin = np.sin(values)
t_sin1 = [taylor(sin(x), 1, 1).subs(x,v) for v in values]
t_sin2 = [taylor(sin(x), 1, 2).subs(x,v) for v in values]
t_sin3 = [taylor(sin(x), 1, 3).subs(x,v) for v in values]
t_sin4 = [taylor(sin(x), 1, 4).subs(x,v) for v in values]
legends = ['sin() ','Taylor 0 (constant)','Taylor 1 (linear)','Taylor 2 (quadratic)','Taylor 3 (cubic)']

fig, ax = plt.subplots()
ax.plot(values,p_sin, color ='red')
ax.plot(values,t_sin1)
ax.plot(values,t_sin2)
ax.plot(values,t_sin3)
ax.plot(values,t_sin4)

ax.set_ylim([-1.3,1.3])
ax.axhline(y=0.0, xmin=-1.0, xmax=1.0, color='black')
ax.axvline(x=0.0, ymin=-1.0, ymax=1.0, color='black')
ax.legend(legends)

plt.show()
```
