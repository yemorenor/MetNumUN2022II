```python
from sympy import Symbol
from sympy import div

x = Symbol('x')

p = x**5 - 19*x**4 + 101*x**3 - 5*x**2 - 1014*x + 1512

p, r = div(p,  x+3)

print(p)
print(r)

p, r = div(p,  x-2)

print(p)
print(r)

p, r = div(p,  x-4)

print(p)
print(r)

p, r = div(p,  x-7)

print(p)
print(r)
```

```python
from sympy import symbols, I, N, pi, expand
x1, y1, z1 = symbols("x1 y1 z1", real=True)  
x = x1 + z1 
y = y1 + z1 
z = z1 + z1 
print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
print(N(pi,4))
w=N(1/(pi + I), 20)
print('w=',w)
```
